```python
response = requests.get(url) 
data = response.json()
arr_with_index = np.random.randint(0, len(data['photos']), size=3)
```

```python
fig, ax = plt.subplots(figsize=(10, 20), nrows=1, ncols=3)
for i in range(len(arr_with_index)):
    url_img = data['photos'][arr_with_index[i]]['img_src']
    img = io.imread(url_img)
    ax[i].imshow(img, cmap='gray')
for axi in ax.ravel(): 
    axi.set_axis_off()
    
plt.show()
```
